```python
# Import Libraries

# First let's import the packages we will use in this project
# You can do this all now or as you need them

import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

#Adjusts the configuration of the plots we will create
%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)
```

```python
# Now we need to read in the data

df = pd.read_csv('/Users/omid/Data/Omid Python/movies.csv')
```

```python
# Let's look at the data

df.head()
```

```python
# We need to see if we have any missing data
# Let's loop through the data and see if there is anything missing

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {:.1f}%'.format(col, pct_missing))
```

```python
#Delete row with dummy value

df = df.dropna(how='any',axis=0)
```

```python
# Data Types for our columns

print(df.dtypes)
```

```python
# change data type of columns

df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
df['votes'] = df['votes'].astype('int64')
```

```python
df.head()
```

```python
# Create correct yearl column

#df['released'] = df['released'].astype(str)
#df['yearCorrect','yearCorrect2'] = df['released'].str.split(',', expand = True)
```

```python
# Drop any duplicates

df.drop_duplicates()
df.head()
```

```python
pd.set_option('display.max_row', None)
```

```python
# Sort by gross

df = df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
# Budget high correlation
# Company high correlation

```

```python
# Scatter plot with budget vs gross

plt.scatter(x=df['budget'], y=df['gross'])
plt.title('Budget vs Gross Earnings')
plt.xlabel('Gross Earnings')
plt.ylabel('Budget for film')
plt.show()
```

```python
df.head()
```

```python
# Plot budget vs gross using seaborn

sns.regplot(x="gross", y="budget", data=df, scatter_kws={'color':'red'}, line_kws={'color':'blue'})
```

```python
# Let's start looking at correlation
# pearson, kendall, spearman

df.corr(method='pearson')
```

```python
df.corr(method='kendall')
```

```python
df.corr(method='spearman')
```

```python
correlation_matrix = df.corr(method='pearson')
sns.heatmap(correlation_matrix, annot=True)

plt.title("Correlation matrix for Numeric Features")
plt.xlabel("Movie features")
plt.ylabel("Movie features")

plt.show()
```

```python
df_numerized = df

for col_name in df_numerized.columns:
    if (df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes

df_numerized.head()
```
